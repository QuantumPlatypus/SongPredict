### Create the training and test feature sets

```python
Combined = numpy.concatenate((X_train, kspace), axis = 1)
C_test = numpy.concatenate((X_test, k_test), axis = 1)
Combined.shape
```

## Fitting the model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10, verbose=True,max_depth=None,min_samples_split=1, random_state=0)
```

```python
from sklearn.metrics import confusion_matrix
```

###Combined data

```python
model.fit(Combined,Y_train)
Y_test_pred = model.predict(C_test)
my_score = model.score(C_test,Y_test)
print my_score
confusion_matrix(Y_test, Y_test_pred)
```

###Time only

```python
model.fit(X_train,Y_train)
Y_test_pred = model.predict(X_test)
my_score = model.score(X_test,Y_test)
print my_score
confusion_matrix(Y_test, Y_test_pred)
```

##K-space only

```python
model.fit(kspace,Y_train)
Y_test_pred = model.predict(k_test)
my_score = model.score(k_test,Y_test)
print my_score
confusion_matrix(Y_test, Y_test_pred)
```
