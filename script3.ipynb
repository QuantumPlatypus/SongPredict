Next lets try short time Fourier transform
    

```python
import scipy, pylab
import numpy as np
def stft(x, framesz, hop):
    """
     x - signal
     fs - sample rate
     framesz - frame size
     hop - hop size (frame size = overlap + hop size)
    """
    framesamp = framesz
    hopsamp = hop
    w = scipy.hamming(framesamp)
    X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X


```

One problem is that the distinctive points are shifting in time.  To pick out distinctive notes/patterns, take the peak power of the short-time fourier transform spectrogram

Variables that can be adjusted:  
    - size of the window   
    - overlap   
    - window in which we search for the local max

This next piece of code just focuses on one sample.

```python
X_stft = stft(X_train[520,:], 330,165)
X_stft.shape
fig2b = matplotlib.pyplot.gcf()
fig2b.set_size_inches(18.5,10.5)
imshow(abs(X_stft), aspect = 'auto')
```

```python
X_pow = numpy.sum(abs(X_stft), axis = 1)
fig2b = matplotlib.pyplot.gcf()
fig2b.set_size_inches(18.5,10.5)
plot(X_pow)
```

Here is how these peaks work - peak 0 is around time point 165.  Each jump shifts by 165, so peak N is at (N\*165 + 165).

## Fitting the model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10, verbose=True,max_depth=None,min_samples_split=1, random_state=0)
```

```python
from sklearn.metrics import confusion_matrix
```

###Combined data

```python
model.fit(Combined,Y_train)
Y_test_pred = model.predict(C_test)
my_score = model.score(C_test,Y_test)
print my_score
confusion_matrix(Y_test, Y_test_pred)
```
